## 六、ndarray的排序

小测验：
使用以上所学numpy的知识，对一个ndarray对象进行选择排序。  

def Sortn(x):  

代码越短越好

```python
n = [5,2,3,6,9]
```

```python
def bubble(n):
    for i in range(len(n) -1):
        for j in range(i+1, len(n)):
            if n[i] > n[j]:
                n[i], n[j] = n[j], n[i]
                
```

```python
bubble(n)
n
```

```python
# 选择排序
def select(n):
    for i in range(len(n)):
        # 选出最小值的索引
        index = np.argmin(n[i:]) + i
        # 把最小值和当前值的位置换一下
        n[i], n[index] = n[index], n[i]
        
```

```python
n = [4, 6,1,0,3]
select(n)
n
```
